### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



Con este codigo buscamos analizar el ingreso_total de ventas de las 4 tiendas de nuestro cliente durante el año 2020 , 2021 , 2022 y 2023.

```python
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
def calcular_ingreso_total(df, nombre_tienda):
    """Calcula el ingreso total de una tienda."""
    ingreso_total = df['Precio'].sum()
    print(f"\n{nombre_tienda}")
    print(f"Ingreso total: ${ingreso_total:,.2f}")
    return ingreso_total



ingreso_total_tienda = calcular_ingreso_total(tienda, "Tienda 1")
ingreso_total_tienda2 = calcular_ingreso_total(tienda2, "Tienda 2")
ingreso_total_tienda3 = calcular_ingreso_total(tienda3, "Tienda 3")
ingreso_total_tienda4 = calcular_ingreso_total(tienda4, "Tienda 4")

#  Opcional: Ingreso total de todas las tiendas
total_ingresos = sum(
    [
        ingreso_total_tienda,
        ingreso_total_tienda2,
        ingreso_total_tienda3,
        ingreso_total_tienda4,
    ]
)
print(f"\nTodas las Tiendas")
print(f"Ingreso total de todas las tiendas: ${total_ingresos:,.2f}")
```

# 2. Ventas por categoría

```python
def calcular_ventas_por_categoria(df, nombre_tienda):
    """Calcula las ventas por categoría para una tienda."""
    print(f"\n--- {nombre_tienda} - Ventas por Categoría ---")
    ventas_por_categoria = (
        df.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
    )
    print(ventas_por_categoria)
    return ventas_por_categoria


calcular_ventas_por_categoria(tienda, "Tienda 1")
calcular_ventas_por_categoria(tienda2, "Tienda 2")
calcular_ventas_por_categoria(tienda3, "Tienda 3")
calcular_ventas_por_categoria(tienda4, "Tienda 4")

#  Opcional: Ventas por categoría en todas las tiendas
tiendas_concatenadas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
print("\n--- Todas las Tiendas - Ventas por Categoría ---")
print(tiendas_concatenadas.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False))
```

```python
import pandas as pd

# (Código para cargar los datos de las tiendas aquí - tienda1, tienda2, tienda3, tienda4)


def verificar_total_ventas_categoria(ventas_categoria, nombre_tienda):
    """
    Verifica que la suma de las ventas por categoría coincida con el ingreso total de la tienda.

    Args:
        ventas_categoria (pd.Series): Serie de Pandas con las ventas por categoría.
        nombre_tienda (str): Nombre de la tienda.

    Returns:
        float: Suma total de las ventas por categoría.
    """
    total_ventas_categoria = ventas_categoria.sum()
    print(f"\n--- Verificación de Ventas por Categoría - {nombre_tienda} ---")
    print(f"Suma de ventas por categoría: ${total_ventas_categoria:,.2f}")

# Calcular y verificar para cada tienda
ventas_categoria_tienda1 = calcular_ventas_por_categoria(tienda, "Tienda 1")
verificar_total_ventas_categoria(ventas_categoria_tienda1, "Tienda 1")

ventas_categoria_tienda2 = calcular_ventas_por_categoria(tienda2, "Tienda 2")
verificar_total_ventas_categoria(ventas_categoria_tienda2, "Tienda 2")

ventas_categoria_tienda3 = calcular_ventas_por_categoria(tienda3, "Tienda 3")
verificar_total_ventas_categoria(ventas_categoria_tienda3, "Tienda 3")

ventas_categoria_tienda4 = calcular_ventas_por_categoria(tienda4, "Tienda 4")
verificar_total_ventas_categoria(ventas_categoria_tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
def calcular_calificacion_promedio(df, nombre_tienda):
    """Calcula la calificación promedio de los clientes para una tienda."""

    calificacion_promedio = df['Calificación'].mean()
    print(f"\n--- {nombre_tienda} ---\nCalificación promedio: {calificacion_promedio:.2f}")
    return calificacion_promedio

# Calcular la calificación promedio para cada tienda
calificacion_promedio_tienda1 = calcular_calificacion_promedio(tienda, "Tienda 1")
calificacion_promedio_tienda2 = calcular_calificacion_promedio(tienda2, "Tienda 2")
calificacion_promedio_tienda3 = calcular_calificacion_promedio(tienda3, "Tienda 3")
calificacion_promedio_tienda4 = calcular_calificacion_promedio(tienda4, "Tienda 4")

# (Opcional) Calcular la calificación promedio de todas las tiendas
calificaciones = pd.concat([tienda['Calificación'], tienda2['Calificación'], tienda3['Calificación'], tienda4['Calificación']])
calificacion_promedio_total = calificaciones.mean()
print(f"\n--- Todas las Tiendas ---\nCalificación promedio: {calificacion_promedio_total:.2f}")
```

# 4. Productos más y menos vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Para visualizaciones más atractivas

def analizar_ventas_productos(df, nombre_tienda):
    """
    Analiza los productos más y menos vendidos en una tienda.

    Args:
        df (pd.DataFrame): DataFrame con los datos de la tienda.
        nombre_tienda (str): Nombre de la tienda para identificarla en la salida.
    """

    # Contar la frecuencia de cada producto
    conteo_productos = df['Producto'].value_counts()

    # Productos más vendidos
    mas_vendidos = conteo_productos.head(5)  # Obtener los 5 más vendidos
    print(f"\n--- {nombre_tienda} - Productos Más Vendidos ---\n{mas_vendidos}\n")

    # Productos menos vendidos
    menos_vendidos = conteo_productos.tail(5)  # Obtener los 5 menos vendidos
    print(f"\n--- {nombre_tienda} - Productos Menos Vendidos ---\n{menos_vendidos}\n")

    # Visualización (Gráfico de barras para los más vendidos)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=mas_vendidos.index, y=mas_vendidos.values)
    plt.title(f"{nombre_tienda} - Productos Más Vendidos")
    plt.xlabel("Producto")
    plt.ylabel("Cantidad de Ventas")
    plt.xticks(rotation=45, ha="right")  # Rotar etiquetas para mejor legibilidad
    plt.tight_layout() #Ajustar el diseño para que las etiquetas no se superpongan
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.barplot(x=menos_vendidos.index, y=menos_vendidos.values)
    plt.title(f"{nombre_tienda} - Productos Menos Vendidos")
    plt.xlabel("Producto")
    plt.ylabel("Cantidad de Ventas")
    plt.xticks(rotation=45, ha="right")  # Rotar etiquetas para mejor legibilidad
    plt.tight_layout()
    plt.show()

# Aplicar el análisis a cada tienda
analizar_ventas_productos(tienda, "Tienda 1")
analizar_ventas_productos(tienda2, "Tienda 2")
analizar_ventas_productos(tienda3, "Tienda 3")
analizar_ventas_productos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
def calcular_promedio_envio(df, nombre_tienda):
    """Calcula el costo promedio de envío para una tienda."""
    promedio_envio = df['Costo de envío'].mean()
    print(f"\n--- {nombre_tienda} ---\nCosto promedio de envío: ${promedio_envio:.2f}")  # Formateado a 2 decimales
    return promedio_envio

# Calcular el promedio de envío para cada tienda
promedio_envio_tienda1 = calcular_promedio_envio(tienda, "Tienda 1")
promedio_envio_tienda2 = calcular_promedio_envio(tienda2, "Tienda 2")
promedio_envio_tienda3 = calcular_promedio_envio(tienda3, "Tienda 3")
promedio_envio_tienda4 = calcular_promedio_envio(tienda4, "Tienda 4")

# (Opcional) Calcular el promedio de envío general
costos_envio_totales = pd.concat([tienda['Costo de envío'], tienda2['Costo de envío'], tienda3['Costo de envío'], tienda4['Costo de envío']])
promedio_envio_general = costos_envio_totales.mean()
print(f"\n--- Todas las Tiendas ---\nCosto promedio de envío general: ${promedio_envio_general:.2f}")
```

#5.1 Envio promedio por tienda y lugar ( analisis adiconal y opcional)

```python
import pandas as pd

# (Los DataFrames tienda, tienda2, tienda3, tienda4 ya están cargados)

def calcular_envio_promedio_por_tienda(df, nombre_tienda, decimales=2):
    """
    Calcula el costo de envío promedio para cada tienda y lo muestra con el número de decimales especificado.

    Args:
        df (pd.DataFrame): DataFrame con los datos de la tienda.
        nombre_tienda (str): Nombre de la tienda para la impresión.
        decimales (int, opcional): Número de decimales a mostrar. Por defecto es 2.

    Returns:
        pd.Series: Serie con el costo de envío promedio por Lugar de Compra.
    """

    envio_promedio_por_tienda = df.groupby('Lugar de Compra')['Costo de envío'].mean().round(decimales)

    print(f"\n--- {nombre_tienda} - Envío Promedio por Tienda ---")

    # Imprimir encabezados con formato para alinear las columnas
    print(f"\n{'Lugar de Compra':<20} {'Costo de envío promedio':>20}")
    print("-" * 41)

    # Formatear la salida con anchos específicos
    for tienda, promedio in envio_promedio_por_tienda.items():
        print(f"{tienda:<20} {promedio:>20.{decimales}f}")

    return envio_promedio_por_tienda

# Calcular el envío promedio para cada tienda
calcular_envio_promedio_por_tienda(tienda, "Tienda 1")
calcular_envio_promedio_por_tienda(tienda2, "Tienda 2")
calcular_envio_promedio_por_tienda(tienda3, "Tienda 3")
calcular_envio_promedio_por_tienda(tienda4, "Tienda 4")

# Opcional: Calcular el envío promedio para todas las tiendas
# Redondear los costos de envío en cada DataFrame antes de concatenar
tienda['Costo de envío'] = tienda['Costo de envío'].round(2)
tienda2['Costo de envío'] = tienda2['Costo de envío'].round(2)
tienda3['Costo de envío'] = tienda3['Costo de envío'].round(2)
tienda4['Costo de envío'] = tienda4['Costo de envío'].round(2)

tiendas_concatenadas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
envio_promedio_todas_tiendas = tiendas_concatenadas.groupby('Lugar de Compra')['Costo de envío'].mean().round(2) # Asegurar el redondeo aquí también

print("\n--- Todas las Tiendas - Envío Promedio por Tienda ---")

# Imprimir encabezados y datos formateados (como en la función)
print(f"\n{'Lugar de Compra':<20} {'Costo de envío promedio':>20}")
print("-" * 41)
for tienda, promedio in envio_promedio_todas_tiendas.items():
    print(f"{tienda:<20} {promedio:>20.2f}")
```

#1. Analisis grafico "Facturacion por Tienda"

```python
import matplotlib.pyplot as plt

# Datos de las tiendas
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [
    1150880400,
    1116343500,
    1098019600,
    1038375700
]

# Crear la figura y los ejes del gráfico usando subplots
# Puede que necesites ajustar el tamaño (figsize) si los textos se solapan
fig, ax = plt.subplots(figsize=(10, 7))

# Crear el gráfico de barras verticales
bars = ax.bar(nombres_tiendas, ingresos, color=['skyblue', 'lightgreen', 'salmon', 'gold'])

# Añadir etiquetas a los ejes y título
ax.set_xlabel('Tienda')
ax.set_ylabel('Ingreso Total ($)')
ax.set_title('Ingreso Total por Tienda')

# --- Establecer el rango del eje Y ---
# Mantenemos el rango en la escala original de los ingresos para que las barras se muestren correctamente
ax.set_ylim(0, 1.4e9) # 1.5 mil millones
# -----------------------------------

# --- Añadir el valor en la parte superior de cada barra en horizontal (formato millones) ---
for bar in bars:
    yval = bar.get_height() # Obtiene la altura de la barra (el valor del ingreso original)
    xval = bar.get_x() + bar.get_width()/2 # Calcula la posición X central de la barra

    # --- Formatear el valor en millones ---
    # 1. Dividir el valor original por 1 millón
    valor_en_millones = yval / 1_000_000
    # 2. Formatear el número (con separador de miles opcional, 0 decimales) y añadir el texto
    texto_a_mostrar = f'{valor_en_millones:,.0f} millones' # Ejemplo: "1,151 millones"
    # Si quieres exactamente 4 dígitos sin coma, podrías usar:
    # texto_a_mostrar = f'{valor_en_millones:.0f} millones' # Ejemplo: "1151 millones"
    # --------------------------------------

    ax.text(
        xval,             # Posición X: centrada horizontalmente en la barra
        yval,             # Posición Y: en la parte superior de la barra
        texto_a_mostrar,  # El texto formateado
        va='bottom',      # Alineación vertical: justo encima de la barra
        ha='center',      # Alineación horizontal: centrar el texto
        # rotation=0,     # El texto es horizontal por defecto
        fontsize=9        # Tamaño de fuente opcional (ajusta si hay solapamiento)
    )

# Ajustar diseño para evitar solapamiento
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
